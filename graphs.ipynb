```python
import numpy as np

processors = np.array([2, 4, 8, 10])
runtime = np.array([7.8e-05, 0.000172, 0.000223, 0.000265])
speedup = np.array([25641.03, 23255.81, 35874.44, 37735.85])
efficiency = np.array([12820.51, 5813.95, 4484.31, 3773.59])
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting runtime on the primary y-axis
ax1.set_xlabel('Processors')
ax1.set_ylabel('Runtime', color='tab:blue')
ax1.plot(processors, runtime, 'o-', color='tab:blue', label='Runtime')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Creating a second y-axis for speedup and efficiency
ax2 = ax1.twinx()
ax2.set_ylabel('Speedup & Efficiency', color='tab:red')
ax2.plot(processors, speedup, 's-', color='tab:orange', label='Speedup')
ax2.plot(processors, efficiency, 'd-', color='tab:red', label='Efficiency')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adding legends and grid
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
ax1.grid()

plt.title('Performance Metrics vs. Processors')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

processors = np.array([2, 4, 8, 10])
runtime = np.array([0.092451, 0.050886, 0.03639, 0.024967])
speedup = np.array([21.64, 78.62, 219.64, 400.53])
efficiency = np.array([10.82, 19.66, 27.46, 40.05])

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting runtime on the primary y-axis
ax1.set_xlabel('Processors')
ax1.set_ylabel('Runtime', color='tab:blue')
ax1.plot(processors, runtime, 'o-', color='tab:blue', label='Runtime')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Creating a second y-axis for speedup and efficiency
ax2 = ax1.twinx()
ax2.set_ylabel('Speedup & Efficiency', color='tab:red')
ax2.plot(processors, speedup, 's-', color='tab:orange', label='Speedup')
ax2.plot(processors, efficiency, 'd-', color='tab:red', label='Efficiency')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adding legends and grid
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
ax1.grid()

plt.title('Performance Metrics vs. Processors for 10000x10000 Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
processors = np.array([2, 4, 8, 10])
T1 = np.array([2.91579, 2.91579, 2.91579, 2.91579])
Tp = np.array([2.27785, 1.80377, 1.44458, 1.31345])
speedup = np.array([1.28, 1.62, 2.02, 2.22])
efficiency = np.array([0.64, 0.41, 0.25, 0.22])

# Plotting
plt.figure(figsize=(10, 6))

# Speedup
plt.plot(processors, speedup, 'o-', label="Speedup", color='blue')

# Efficiency
plt.plot(processors, efficiency, 's-', label="Efficiency", color='red')

plt.xlabel('Number of Processors (p)')
plt.ylabel('Value')
plt.title('Speedup and Efficiency vs. Number of Processors')
plt.xticks(processors)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
